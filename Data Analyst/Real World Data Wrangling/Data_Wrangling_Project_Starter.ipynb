Clean the data to solve the 4 issues corresponding to data quality and tidiness found in the assessing step. **Make sure you include justifications for your cleaning decisions.**

First we need to restructure the dataset before clean data quality issues.

```python
# FILL IN - Make copies of the datasets to ensure the raw dataframes 
# are not impacted
df_first_cleaned = df_first.copy()
df_first_cleaned.head(5)
```

```python
df_last_cleaned = df_last.copy()
df_last_cleaned.head(5)
```

```python
df_company_info_cleaned = df_company_info.copy()
df_company_info_cleaned
```

### **Tidiness Issue 1: FILL IN**

```python
#FILL IN - Apply the cleaning strategy

#CLEANING 2 PRICE DATASETS

##skip unvaluable 2 rows
df_first_cleaned = df_first_cleaned.iloc[2:]
df_last_cleaned = df_last_cleaned.iloc[2:]

##First day data: with first_day dataset we need to keep 4 variables: symbol, open_price, percent_change.
###rename columns to be more meaningful.
df_first_cleaned = df_first_cleaned.rename(columns = {'Unnamed: 1':'symbol','Unnamed: 5':'open_price', \
                                                      'Unnamed: 7':'change','Unnamed: 8':'percent_change'})
df_first_cleaned = df_first_cleaned.drop(columns=['Unnamed: 0','Unnamed: 2','Unnamed: 3','Unnamed: 4','Unnamed: 6', \
                       'Unnamed: 9','Unnamed: 10','Unnamed: 11','Unnamed: 12','Unnamed: 13','Unnamed: 14', \
                       'Unnamed: 15','Unnamed: 16','Unnamed: 17'])

##Last day data: with first_day dataset we need to keep 2 variables: symbol, close_price.
df_last_cleaned = df_last_cleaned.rename(columns = {'Unnamed: 1':'symbol','Unnamed: 6':'close_price'})
df_last_cleaned = df_last_cleaned.drop(columns=['Unnamed: 0','Unnamed: 2','Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 5', \
                       'Unnamed: 9','Unnamed: 10','Unnamed: 11','Unnamed: 12','Unnamed: 13','Unnamed: 14', \
                       'Unnamed: 7','Unnamed: 8','Unnamed: 15','Unnamed: 16','Unnamed: 17'])
```

```python
#FILL IN - Validate the cleaning was successful
df_first_cleaned.head()
```

```python
df_last_cleaned.head()
```

### **Tidiness Issue 2: FILL IN**

```python
#FILL IN - Apply the cleaning strategy
#We need to merge df_first_cleaned, df_last_cleaned into one dataframe for analysis
df_prices = pd.merge(df_first_cleaned,df_last_cleaned,how='inner',on='symbol')
```

```python
#FILL IN - Validate the cleaning was successful
df_prices.head(5)
```

### **Quality Issue 1: FILL IN**

```python
# FILL IN - Apply the cleaning strategy
df_prices['open_price'] = df_prices['open_price'].astype(float)
df_prices['change'] = df_prices['change'].astype(float)
df_prices['percent_change'] = df_prices['percent_change'].astype(float)
df_prices['close_price'] = df_prices['close_price'].astype(float)
```

```python
# FILL IN - Validate the cleaning was successful
df_prices.info()
```

### **Quality Issue 2: FILL IN**

```python
#FILL IN - Apply the cleaning strategy
#Remove invalid, we just need to keep rows have type = s (stock)
df_company_info_cleaned = df_company_info_cleaned[df_company_info_cleaned['type'] == 's']
```

```python
#FILL IN - Validate the cleaning was successful
df_company_info_cleaned.info()
```

### **Remove unnecessary variables and combine datasets**

Depending on the datasets, you can also peform the combination before the cleaning steps.

```python
df_prices
```

```python
#FILL IN - Remove unnecessary variables and combine datasets
#Merge merge price data and company info data
df_company_info_cleaned = df_company_info_cleaned.rename(columns = {'code':'symbol'})
df_prices = pd.merge(df_prices,df_company_info_cleaned,how='left',on='symbol')

#Remove unneccessary columns
df_prices = df_prices.drop(columns=['clientName','securityName','stockNo','description','full_name','name','sort','symbol_ref','lastDateSyncIDS','createdAt','updatedAt'])


#calculate change and percent change
df_prices['change'] = df_prices['close_price'] - df_prices['open_price']
df_prices['percent_change'] = df_prices['change'] / df_prices['open_price'] * 100
df_prices
```

## 5. Answer the research question

### **5.1:** Define and answer the research question 
Going back to the problem statement in step 1, use the cleaned data to answer the question you raised. Produce **at least** two visualizations using the cleaned data and explain how they help you answer the question.

*Research question:* 
- Top 10 companies with the highest growth rate in the first 9 months of 2023.
- Top 10 companues have highest market capitalization.

```python
#Visual 1 - FILL IN
import matplotlib.pyplot as plt

#filter data by exchange
top10_highest_growth_hose = df_prices[df_prices['exchange']=='HOSE']
top10_highest_growth_hose = top10_highest_growth_hose.sort_values(by='percent_change', ascending=False).head(20)
top10_highest_growth_hnx = df_prices[df_prices['exchange']=='HNX']
top10_highest_growth_hnx = top10_highest_growth_hnx.sort_values(by='percent_change', ascending=False).head(20)
top10_highest_growth_upcom = df_prices[df_prices['exchange']=='UPCOM']
top10_highest_growth_upcom = top10_highest_growth_upcom.sort_values(by='percent_change', ascending=False).head(20)

fig, axes = plt.subplots(3,1, figsize=(10,10))
ax1 = axes[0];
ax2 = axes[1];
ax3 = axes[2];

ax1.bar(top10_highest_growth_hose['symbol'],top10_highest_growth_hose['percent_change'])
ax1.set_title('HOSE')

ax2.bar(top10_highest_growth_hnx['symbol'],top10_highest_growth_hnx['percent_change'])
ax2.set_title('HNX')

ax3.bar(top10_highest_growth_upcom['symbol'],top10_highest_growth_upcom['percent_change'])
ax3.set_title('UPCOM')
```

*Answer to research question:*
Top-performing companies with the highest growth rate in the HOSE, HNX, and UPCOM exchanges in the first 9 months of 2023.
- HOSE: QCG
- HXN: CMS
- UPCOM: CH5

```python
#Visual 2 - FILL IN
import matplotlib

top10_highest_growth = df_prices.sort_values(by='marketCapPrev', ascending=False).head(10)
top10_highest_growth.set_index('symbol',inplace=True)
top10_highest_growth['marketCapPrev'].plot(kind='bar',figsize=(10,6))
```
